# EXPLORATORY DATA ANALYSIS

## 3.1 HYPOTHESIS TESTING VERSUS EXPLORATORY DATA ANALYSIS

When approaching a data mining problem, a data mining analyst may already have some a priori hypotheses that he or she would like to test regarding the relationships between the variables.

However, analysts do not always have a priori notions of the expected relationships among the variables. Especially when confronted with unknown, large databases, analysts often prefer to use exploratory data analysis (EDA), or graphical data analysis. EDA allows the analyst to
* Delve into the data set;
* Examine the inter-relationships among the attributes;
* Identify interesting subsets of the observations; and 
* Develop an initial idea of possible associations among the predictors, as well as between the predictors and the target variable.
