```python
import scipy.stats as stats
import numpy as np 

# Define your data as a list of lists, where each inner list represents a group

# Group A, B, C = data[0], data[1], data[2]
data = np.array([[30, 40, 50, 60],
    [25, 30, 50, 55],
    [25, 30, 40, 45]])

# Perform the ANOVA test
f_stat, p_val = stats.f_oneway(*data)

# Print the results
print(f"F-statistic: {np.round(f_stat,2)}")
print(f"p-value: {np.round(p_val,2)}")
```

```python
import numpy as np
import pandas as pd
import scipy.stats as st
import matplotlib.pyplot as plt

input_file = "./data_sets/cereals.CSV"

df = pd.read_csv(input_file)
df = df[['Sugars','Rating']]
df.dropna(inplace=True)

X = df['Sugars'] # independent variable
Y = df['Rating'] # dependent variable

# Perform a linear regression using scipy
slope, intercept, r_value, p_value, std_err = stats.linregress(X, Y)

# Print the regression results
print("Slope:", np.round(slope,2))
print("Intercept:", np.round(intercept,2))
print("R-squared:", np.round(r_value**2,2))
print("p-value:", p_value)

df.plot(x='Sugars', y='Rating', kind='scatter',s=50,c="white",edgecolor="red")
xx = np.linspace(X.min(),X.max(),100)
yy = slope*xx+intercept
plt.plot(xx,yy)
plt.show()
```

```python
import numpy as np
import pandas as pd
import scipy.stats as st
import matplotlib.pyplot as plt

input_file = "./data_sets/cereals.CSV"

df = pd.read_csv(input_file)
df = df[['Sugars','Rating']]
df.dropna(inplace=True)

X = df['Sugars'] # independent variable
Y = df['Rating'] # dependent variable

# Perform a linear regression using scipy
slope, intercept, r_value, p_value, std_err = st.linregress(X, Y)

# Print the regression results
print("Slope:", np.round(slope,2))
print("Intercept:", np.round(intercept,2))
print("R-squared:", np.round(r_value**2,2))
print("p-value:", p_value)

df.plot(x='Sugars', y='Rating', kind='scatter',s=50,c="white",edgecolor="red")
xx = np.linspace(X.min(),X.max(),100)
yy = slope*xx+intercept
plt.plot(xx,yy)
plt.show()
```

```python
import statsmodels.api as sm
# Fit linear regression model
model = sm.OLS(Y, sm.add_constant(X)).fit()

# calculate confidence interval for predictions
conf_int = model.get_prediction(sm.add_constant(X)).conf_int()

# calculate prediction interval for individual observations
pred_int = model.get_prediction(sm.add_constant(X)).summary_frame(alpha=0.05)

# print the confidence interval and prediction interval dataframes
# print("Confidence Interval:\n", conf_int)
# print("\nPrediction Interval:\n", pred_int)

fig,ax = plt.subplots(1,1,figsize=(10,7))
# plot data points
plt.scatter(X, Y)

# plot regression line
plt.plot(X, model.fittedvalues)

# plot confidence interval
plt.fill_between(X, conf_int[:, 0], conf_int[:, 1], alpha=0.2)

# plot prediction interval
plt.fill_between(X, pred_int['obs_ci_lower'], pred_int['obs_ci_upper'], alpha=0.2)

# set plot labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression with Confidence and Prediction Intervals')
plt.show()

# Print summary of fit
print(model.summary())
```

```python
df = pd.read_csv(input_file)
df = df[['Sugars','Sodium','Rating']]
df.dropna(inplace=True)

X = df[['Sugars','Sodium']] # independent variable
Y = df['Rating'] # dependent variable
model = sm.OLS(Y, sm.add_constant(X)).fit()
print(model.summary())

new_x = np.array([[1, 1, 290]])
pred = model.get_prediction(new_x)

# print predicted value of y and its prediction interval
print(f"Predicted value of y for x1 = {new_x[0,1]} and x2 = {new_x[0,2]}:", pred.predicted_mean)

# calculate confidence interval for predictions
print("95% prediction interval:", pred.conf_int(alpha=0.05))

# calculate prediction interval for individual observations
pred_int = pred.summary_frame(alpha=0.05)
pred_int
```

```python
# plot 1 - Normal QQ plot of residuals
fig = sm.qqplot(model.resid, line='r')
plt.title("Normal QQ plot of residuals")
plt.show()
```

```python
# plot 2 - Residuals vs Fitted values
fig = plt.figure(figsize=(10,7))
sm.graphics.plot_regress_exog(model, 1, fig=fig)
plt.show()

fig = plt.figure(figsize=(10,7))
sm.graphics.plot_regress_exog(model, 2, fig=fig)
plt.show()
```
